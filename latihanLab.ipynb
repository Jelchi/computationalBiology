***Sequence Manipulation***

```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO

# import sequence
fasta_a = SeqIO.read("seq1.fasta", "fasta")
fasta_b = SeqIO.read("seq2.fasta", "fasta")

seqA = fasta_a.seq
seqB = fasta_b.seq

# Checking Squenece
print("Sequence A: ", seqA)
print("Sequence B: ", seqB)
```

```python
# Total length of sequence
print("Length of SeqA: ", len(seqA))
print("Length of SeqB: ", len(seqB))
```

```python
# Count how many times (CAG) appear
count_CAG_seqA = seqA.count('CAG')
count_CAG_seqB = seqB.count('CAG')
print("CAG in seqA: ", count_CAG_seqA)
print("CAG in seqB: ",count_CAG_seqB)
```

```python
# Find the fist Index (TCA)
find_first_TCA_seqA = seqA.find('TCA')
find_first_TCA_seqB = seqB.find('TCA')
print("The 1st idx (TCA) at seqA :", find_first_TCA_seqA)
print("The 1st idx (TCA) at seqB :", find_first_TCA_seqB) 

# Find the last Index (TCA)
find_last_TCA_seqA = seqA.rfind('TCA')
find_last_TCA_seqB = seqB.rfind('TCA')
print("The last idx (TCA) at seqA: ", find_last_TCA_seqA)
print("The last idx (TCA) at seqB: ", find_last_TCA_seqB)
```

```python
from Bio.SeqUtils import nt_search

# Use nt_search library
print(nt_search(str(seqA), "TCA"))
print(nt_search(str(seqB), "TCA"))
```

```python
# Take the last 10 of seqA and the first 5 of SeqB. Combine into new sequence called seqC
seqC = seqA[-10:] + seqB[:5] # Slicing & Join
print("SeqC: ",seqC)
```

```python
# Reverse seqC
print("reverse seqC:", seqC[::-1])
```

***Sequence Analysis & Plotting***

```python
from collections import Counter

# Find 'C', 'G', 'A', 'T' frequency
seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

print(seqA_freq)
print(seqB_freq)
print(seqC_freq)
```

```python
import matplotlib.pyplot as plt

# Display frequency Diagram 
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.title("Frequency of Seq A")
plt.show()

plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.title("Frequency of SeqB")
plt.show()

plt.bar(seqC_freq.keys(), seqC_freq.values())
plt.title("Frequency of SeqB")
plt.show()
```

```python
from Bio.SeqUtils import gc_fraction, MeltingTemp as mt, molecular_weight
# Find GC & AT content percentage 
gc_a = gc_fraction(seqA) * 100
gc_b = gc_fraction(seqB) * 100
gc_c = gc_fraction(seqC) * 100

print("GC_Fraction Seq A:", gc_a)
print("GC_Fraction Seq B:", gc_b)
print("GC_Fraction Seq C:", gc_c)
```

```python
at_a = 100 - gc_a
at_b = 100 - gc_b
at_c = 100 - gc_c

print("AT_fraction Seq A:", at_a)
print("AT_fraction Seq B:", at_b)
print("AT_fraction Seq C:", at_c)
```

```python
# Wallace Method
mt_A = mt.Tm_Wallace(seqA)
mt_B = mt.Tm_Wallace(seqB)
mt_C = mt.Tm_Wallace(seqC)

print("Melting Temperature of Seq A:", mt_A)
print("Melting Temperature of Seq B:",mt_B)
print("Melting Temperature of Seq C:", mt_C)
```

```python
# Salt Correction Method
sc_A = mt.Tm_GC(seqA)
sc_B = mt.Tm_GC(seqB)
sc_C = mt.Tm_GC(seqC)

print("Melting Temperature of Seq A:", sc_A)
print("Melting Temperature of Seq B:", sc_B)
print("Melting Temperature of Seq C:", sc_C)
```

```python
# Nearest Neighbour
nn_seqA = mt.Tm_NN(seqA)
nn_seqB = mt.Tm_NN(seqB)
nn_seqC = mt.Tm_NN(seqC)

print("Melting Temperature of Seq A:", nn_seqA)
print("Melting Temperature of Seq B:", nn_seqB)
print("Melting Temperature of Seq C:", nn_seqC)
```

```python
# Molecular Weight
print("Molecular Weigth of Seq A:", molecular_weight(seqA))
print("Molecular Weigth of Seq B:", molecular_weight(seqB))
print("Molecular Weigth of Seq C:", molecular_weight(seqC))
```

***DNA & mRNA protein Synthesis***

```python
# Transcribe DNA into mRNA
mRNA_seqA = seqA.transcribe()
mRNA_seqB = seqB.transcribe()
mRNA_seqC = seqC.transcribe()

print("mRNA seq A:", mRNA_seqA)
print("mRNA seq B:", mRNA_seqB)
print("mRNA seq C:", mRNA_seqC)
```

```python
# Translate mRNA into amino acids
amino_acid_seqA = mRNA_seqA.translate()
amino_acid_seqB = mRNA_seqB.translate()
amino_acid_seqC = mRNA_seqC.translate()

print("Amino acid of seqA: ", amino_acid_seqA)
print("Amino acid of seqB:", amino_acid_seqB)
print("Amino acid of seqC", amino_acid_seqC)
```

***Sequence Alignment & Similarities***

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

seqX = seqA[:50]
seqY = seqB[:50]

seq1 = Seq("GCACGACTGCGTGATATTACCGGA")
seq2 = Seq("TATACGTGGACAGGCTCTGGAG")
```

```python
alignment1 = pairwise2.align.localxx(seqX, seq1)
alignment2 = pairwise2.align.localxx(seqY, seq2)
```

```python
for align in alignment1: 
    print(format_alignment(*align))
```

```python
for align in alignment2: 
    print(format_alignment(*align))
```

```python
# Levenshtein Distance
from Levenshtein import distance

distances = distance(seqX, seqY)
print("Levenshtein distance:", distances)
```
